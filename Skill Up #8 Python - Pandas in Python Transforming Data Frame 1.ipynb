```python
## Split data frame
split = [0, 300, 600, len(df_titanic)]
split
```

```python
## Looping to save splited data frame
for i in range(len(split)-1):
    df = df_titanic[split[i] : split[i+1]]
    df.to_csv('df_titanic' + str(i + 1) + '.csv', index=False)
```

## Null Checker

```python
df_titanic.isna().sum()
```

```python
## Fill Age Null with median
df_titanic['Age'] = df_titanic['Age'].fillna(df_titanic['Age'].median())
```

```python
df_titanic.isna().sum()
```

```python
## Fill Embarked Null with modus
df_titanic['Embarked'] = df_titanic['Embarked'].fillna(df_titanic['Embarked'].mode()[0])
```
